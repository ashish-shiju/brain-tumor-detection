```
arr = Y_arr + N_arr
```

```
def load_image(img_arr, img_size):
    X = []
    y = []
    img_width, img_height = img_size
    count = 1

    for filename in arr:
        image = cv.imread(filename)
        image = cv.resize(image, dsize=(img_width, img_height), interpolation=cv.INTER_CUBIC)
        image = image / 255.
        X.append(image)
        if count <= 155:
            y.append([1])
        else:
            y.append([0])
        count += 1

    X = np.array(X)
    y = np.array(y)
    X, y = sk.utils.shuffle(X, y)

    print(f'Number of examples is: {len(X)}')
    print(f'X shape is: {X.shape}')
    print(f'y shape is: {y.shape}')

    return X, y
```

```
X, y = load_image(arr, (240, 240))
```

```
X_train, X_test, y_train, y_test = sk.model_selection.train_test_split(X, y, test_size=0.15, random_state=42)
```

```
def final_model():
    X_input = keras.layers.Input((240, 240, 3))
    X = keras.layers.ZeroPadding2D((2, 2))(X_input)

    X = keras.layers.Conv2D(32, (7, 7), strides = (1, 1))(X)
    X = keras.layers.BatchNormalization(axis = 3, name = 'bn0')(X)
    X = keras.layers.Activation('relu')(X)

    X = keras.layers.MaxPooling2D((4, 4))(X)
    X = keras.layers.MaxPooling2D((4, 4))(X)
    X = keras.layers.Flatten()(X)
    X = keras.layers.Dense(1, activation='sigmoid')(X)
    model = keras.models.Model(inputs = X_input, outputs = X)

    return model
```

```
model = final_model()
```

```
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x=X_train, y=y_train, batch_size=25, epochs=9, validation_data=(X_test, y_test))
```

```
def test_model(img_name):
    x = []
    image = cv.imread(img_name)
    image = cv.resize(image, dsize=(240, 240), interpolation=cv.INTER_CUBIC)
    image = image / 255.
    x.append(image)
    x_test = np.array(x)
    y_preds = model.predict(x_test)
    if y_preds >= 0.5:
        print("Tumor detected")
    else:
        print("Tumor not detected")
    import matplotlib.image as mpimg
    plt.imshow(mpimg.imread(img_name))
```
